# Correlation exercises

In these exercises you will investigate correlations between variables of the diamonds dataset from the “tidyverse” packages. Do you remember the diamonds dataset? If not try to type `?diamonds`

To make the plots run faster, we will only look at a sample of the dataset, which we will call `dia`. Thus, in all the exercises you should use only the `dia` dataset. We will also adjust plot size, which leave us with the following setup:

```r
library(tidyverse)
options(repr.plot.width=8, repr.plot.height=5)
set.seed(9876)
dia <- diamonds[sample(1:nrow(diamonds), 2500), ]
```

### *Exercise 1*

Make a scatterplot with `carat` as the `x` variable and `price` as the `y` variable using ggplot.

- What type of correlation do you see wrt. direction, shape, strength, and outliers?
- What is the correlation coefficient between the two variables?
- Is there any causal relationship between the two variables and if so which one?

```r
ggplot(dia, aes(carat,price)) +
geom_point()
```
